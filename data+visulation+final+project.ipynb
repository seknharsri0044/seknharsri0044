```python
#4.	Filter the dataframe with only this columns ['continent','location','date','total_cases','total_deaths','gdp_per_ca
#pita','human_development_index'] and update the data frame.
new_dataframe=df.filter(items=['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'])
new_dataframe
```

```python
#a.	Remove all duplicates observations
new_dataframe.drop_duplicates(inplace = True)
new_dataframe
```

```python
#b.	Find missing values in all columns
new_dataframe.isnull()
```

```python
#c.	Remove all observations where continent column value is missing
new_dataframe.dropna(axis = 1)
```

```python
#d.	Fill all missing values with 0
new_dataframe.fillna(0).head(5)
```

```python
#a.	Convert date column in datetime format using pandas.to_datetime
new_dataframe['date'] = pd.to_datetime(df['date'])
new_dataframe
```

```python
#b.	Create new column month after extracting month data from column date.
new_dataframe['month'] = pd.DatetimeIndex(df['date']).month
new_dataframe.head()
```

```python
#a.	Find max value in all columns using groupby function on 'continent colum Tip: use reset_index() after applying groupby

df_grouped = new_dataframe.groupby(['continent']).max()
df_grouped
```

```python
#b.	Store the result in a new dataframe named 'df_groupby'. (Use df_groupby dataframe for all further analysis)
df_groupby==df_grouped  
df_groupby
```

```python
#a.	Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'
df_groupby['ratio'] = df_groupby['total_deaths'] / df_groupby['total_cases'] 
df_groupby['ratio']
```
